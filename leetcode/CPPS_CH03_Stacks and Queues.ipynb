## Q1 Trapping Rain Water * * * * * *
[leetcode:42. Trapping Rain Water](https://leetcode.com/problems/trapping-rain-water/)<br>
Given n non-negative integers representing an elevation map where the width of each bar is 1, compute how much water it is able to trap after raining. <br><br>
The above elevation map is represented by array [0,1,0,2,1,0,1,3,2,1,2,1]. In this case, 6 units of rain water (blue section) are being trapped. Thanks Marcos for contributing this image!<br>

Example:
Input: [0,1,0,2,1,0,1,3,2,1,2,1]<br>
Output: 6<br>


```python
class Solution:
    def trap(self, height) :
        '''
        :type height: List[int]
        :rtype: int
        time: O(N)
        space: O(N)
        '''
        lenHeight = len(height)
        if lenHeight < 2: return 0
        
        leftMax, rightMax = height[:], height[:]
        
        for i in range(1, lenHeight):
            leftMax[i] = [height[i], leftMax[i-1]][height[i] <= leftMax[i-1]]

        for i in range(lenHeight - 2, -1, -1):
            rightMax[i] = [height[i], rightMax[i+1]][height[i] <= rightMax[i+1]]
        
        result = 0 
        for i in range(lenHeight):
            result += min(leftMax[i], rightMax[i])-height[i]
            
        return result
```

```python
test_list = [0,1,0,2,1,0,1,3,2,1,2,1]

s = Solution()
s.trap(test_list)
```

## Q2 Valid Parentheses * *
[leetcode:20. Valid Parentheses](https://leetcode.com/problems/valid-parentheses/)<br>
Given a string containing just the characters '(', ')', '{', '}', '[' and ']', determine if the input string is valid.<br>
<br>
An input string is valid if:

Open brackets must be closed by the same type of brackets.<br>
Open brackets must be closed in the correct order.<br>
Note that an empty string is also considered valid.<br>
<br>
__Example 1__:<br>
Input: "()"<br>
Output: true<br><br>
__Example 2__:<br>
Input: "()[]{}"<br>
Output: true<br><br>
__Example 3__:<br>
Input: "(]"<br>
Output: false<br><br>
__Example 4:__<br>
Input: "([)]"<br>
Output: false<br><br>
__Example 5:__<br>
Input: "{[]}"<br>
Output: true<br>

```python
class Solution:
    def isValid(self, s):
        '''
        :type s: str
        :rtype: bool
        time: O(N) 
        space: O(N)
        '''
        d = {'(':')', '[':']', '{':'}'}
        stack = []
        for i in s:
            if i in d:
                stack.append(i)
            else:
                if stack == []: return False # 비어있는지 체크해줘야함!!!
                tmp = stack.pop()
                if d[tmp] != i:
                    return False
        if len(stack) == 0:
            return True
        else: return False
```

## Q4. Basic Calculator * * * * * *
[leetcode: 224. Basic Calculator](https://leetcode.com/problems/basic-calculator/) <br>
Implement a basic calculator to evaluate a simple expression string.<br>

The expression string may contain open ( and closing parentheses ), the plus + or minus sign -, non-negative integers and empty spaces .<br>
<br>
__Example 1:__
<br>
Input: "1 + 1"<br>
Output: 2<br><br>
__Example 2:__<br>
Input: " 2-1 + 2 "<br>
Output: 3<br><br>
__Example 3:__<br>
Input: "(1+(4+5+2)-3)+(6+8)"<br>
Output: 23<br><br>
Note:<br>
You may assume that the given expression is always valid.<br>
Do not use the eval built-in library function.

```python
class Solution:
    def calculate(self, s):
        '''
        :type s: str
        :rtype: int
        time: O(N) 
        space: O(N)
        '''
        lenS = len(s)
        if lenS < 2: return s
        stack = []
        
        for i in range(lenS-1, -1, -1):
            #+, - 인 경우 stack에 추가 
            if s[i] in {'+', '-', ')'}:
                stack.append(s[i])
                
            # ) 인 경우 괄호 안을 계산 
            elif s[i] == '(':
                tmp, tmp_sign = 0, ''
                while (stack[-1] != ')'): 
                    if stack[-1]== '(': return False 
                    tmp, tmp_sign = self.cal(tmp, tmp_sign, stack.pop())
                if stack.pop() != ')': return False
                stack.append(tmp)
                
            # 숫자일 경우 이전 문자가 숫자이면 합쳐서 stack에 추가, 아니면 그냥 추가 
            elif s[i].isdecimal(): 
                tmp = ''
                while(stack != [] and stack[-1].isdecimal()):
                    tmp = stack.pop() + tmp
                stack.append(s[i]+tmp)
        tmp, tmp_sign = 0, ''
        
        while (stack): 
            tmp, tmp_sign = self.cal(tmp, tmp_sign, stack.pop())
        return str(tmp)
    
    def cal(self, tmp, tmp_sign, p):
        if p in {'+', '-'}:
            tmp_sign = p
        else: 
            if tmp_sign == '+':
                tmp += int(p)
            elif tmp_sign == '-':
                tmp -= int(p)
            else:
                tmp = int(p)
        return tmp, tmp_sign 
```

## Q5. Decode String * * * * 
[leetcode:394. Decode String](https://leetcode.com/problems/decode-string/) <br>
Given an encoded string, return its decoded string.<br>

The encoding rule is: k[encoded_string], where the encoded_string inside the square brackets is being repeated exactly k times. Note that k is guaranteed to be a positive integer.<br>

You may assume that the input string is always valid; No extra white spaces, square brackets are well-formed, etc.<br>

Furthermore, you may assume that the original data does not contain any digits and that digits are only for those repeat numbers, k. For example, there won't be input like 3a or 2[4].<br><br>
Examples:
<br>
s = "3[a]2[bc]", return "aaabcbc".<br>
s = "3[a2[c]]", return "accaccacc".<br>
s = "2[abc]3[cd]ef", return "abcabccdcdcdef".<br>

```python
class Solution:
    def decodeString(self, s):
        '''
        :type s: str
        :rtype str
        time: O(N)
        space: O(N)
        '''
        stack = []
        for i in s:
            if i.isdecimal():
                if (stack != [] and stack[-1].isdecimal()):
                    stack.append(stack.pop()+i)
                else:
                    stack.append(i)
            elif i == '[':
                stack.append(i)
            elif i == ']':
                tmp = ''
                while (stack != [] and stack[-1] != '['):
                    tmp = stack.pop()+tmp
                if ((stack == []) or (stack.pop() != '[')): return False
                if stack != []: tmpNum = stack.pop()
                if tmpNum.isdecimal() == False: return 
                stack.append(int(tmpNum)*tmp)
            else:
                stack.append(i)
        result = ''
        while(stack):
            result = stack.pop() + result
        return result
                
```

## Q6. Stack Min * * * *
[leetcode: 155. Min Stack](https://leetcode.com/problems/min-stack/)<br>
Design a stack that supports push, pop, top, and retrieving the minimum element in constant time.<br>

push(x) -- Push element x onto stack.<br>
pop() -- Removes the element on top of the stack.<br>
top() -- Get the top element.<br>
getMin() -- Retrieve the minimum element in the stack.<br>
 
__Example:__
<br>MinStack minStack = new MinStack();
<br>minStack.push(-2);
<br>minStack.push(0);
<br>minStack.push(-3);
<br>minStack.getMin();   --> Returns -3.
<br>minStack.pop();
<br>minStack.top();      --> Returns 0.
<br>minStack.getMin();   --> Returns -2. 

```python
class MinStack:

    def __init__(self):
        """
        initialize your data structure here.
        """
        self.stack = []
        self.stackMin = []
        
    def push(self, x):
        '''
        :type x:int
        time: O(1)
        '''
        self.stack.append(x)
        if self.stackMin == [] or (self.stackMin[-1] > x):
            self.stackMin.append(x)
        else: self.stackMin.append(self.stackMin[-1])
        
    def pop(self):
        '''
        time: O(1)
        '''
        self.stack.pop()
        self.stackMin.pop()
         
    def top(self):
        '''
        :rtype: int
        time: O(1)
        '''
        return self.stack[-1]

    def getMin(self):
        '''
        :rtype: int
        time: O(1)
        '''
        return self.stackMin[-1]

# Your MinStack object will be instantiated and called as such:
# obj = MinStack()
# obj.push(x)
# obj.pop()
# param_3 = obj.top()
# param_4 = obj.getMin()
```

## Q7. Queue via Stacks * * * *
[leetcode: 232. Implement Queue using Stacks](https://leetcode.com/problems/implement-queue-using-stacks/)<br>

Implement the following operations of a queue using stacks.<br>

push(x) -- Push element x to the back of queue.<br>
pop() -- Removes the element from in front of queue.<br>
peek() -- Get the front element.<br>
empty() -- Return whether the queue is empty.<br><br>
__Example:__<br>
MyQueue queue = new MyQueue();<br><br>
queue.push(1);<br>
queue.push(2);  <br>
queue.peek();  // returns 1<br>
queue.pop();   // returns 1<br>
queue.empty(); // returns false<br>
Notes:<br>

You must use only standard operations of a stack -- which means only push to top, peek/pop from top, size, and is empty operations are valid.<br>
Depending on your language, stack may not be supported natively. You may simulate a stack by using a list or deque (double-ended queue), as long as you use only standard operations of a stack.<br>
You may assume that all operations are valid (for example, no pop or peek operations will be called on an empty queue).

```python
# 통과하긴 했으나 이건 stack을 이용해서 queue를 구현한 것이 아니라고 함. 
class MyQueue:

    def __init__(self):
        """
        Initialize your data structure here.
        """
        self.queue = []
        self.n = 0
        
    def push(self, x):
        """
        :type x: int
        Push element x to the back of queue.
        """
        self.queue.append(x)
        self.n += 1
        
    def pop(self):
        """
        :rtype: int
        Removes the element from in front of queue and returns that element.
        """
        if self.n == 0: return False
        self.n -= 1
        return self.queue.pop(0)
        
    def peek(self):
        """
        :rtype: int
        Get the front element.
        """
        if self.n == 0 : return False
        return self.queue[0]

    def empty(self):
        """
        :rtype: bool
        Returns whether the queue is empty.
        """
        return True if self.n == 0 else False
        


# Your MyQueue object will be instantiated and called as such:
# obj = MyQueue()
# obj.push(x)
# param_2 = obj.pop()
# param_3 = obj.peek()
# param_4 = obj.empty()
```

```python
# 문제의 의도와 같이 stack을 이용해서 queue를 구현해보자. 
class MyQueue:

    def __init__(self):
        """
        Initialize your data structure here.
        """
        self.stack1 = []
        self.stack2 = []
        
    def push(self, x):
        """
        :type x: int
        Push element x to the back of queue.
        timp: O(1)
        """
        self.stack1.append(x)

    def pop(self):
        """
        :rtype: int
        Removes the element from in front of queue and returns that element.
        timp: O(N)
        """
        if self.stack2 == []:
            while self.stack1:
                self.stack2.append(self.stack1.pop())
        return self.stack2.pop()
    
    def peek(self):
        """
        :rtype: int
        Get the front element.
        timp: O(N)
        """
        if self.stack2 == []:
            while self.stack1:
                self.stack2.append(self.stack1.pop())
        return self.stack2[-1]

    def empty(self):
        """
        :rtype: bool
        Returns whether the queue is empty.
        timp: O(1)
        """
        return not self.stack1 and not self.stack2


# Your MyQueue object will be instantiated and called as such:
# obj = MyQueue()
# obj.push(x)
# param_2 = obj.pop()
# param_3 = obj.peek()
# param_4 = obj.empty()
```
